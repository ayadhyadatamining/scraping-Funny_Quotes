# function to  scrape 1st page

```python
import pandas as pd
def quotes(soup1):
    author_name = [i.text.strip() for i in soup1.find_all('span',class_='authorOrTitle')]
    quotes=[list(i)[0].strip() for i in soup1.find_all('div',class_='quoteText')]
    likes=[int(list(i)[0].strip().split()[0]) for i in soup1.find_all('a',class_='smallText')]
    return pd.DataFrame({"Authors":author_name,"Quotes":quotes,"likes":likes})
```

```python
df=quotes(soup1)
df
```
